```python
#딥러닝 관련 라이브러리 불러오기
import tensorflow as tf
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt
```

```python
#데이터셋 불러오기

mnist = keras.datasets.mnist
(x_train, y_train), (x_test, y_test) = mnist.load_data()
x_train.shape
print(len(x_train)) #784
print(len(x_test))
print(28*28)
y_train[0]
```

```python
#데이터 전처리 과정 (0-1까지 숫자로 변경)
x_train = x_train / 255
x_test = x_test / 255

```

```python
#모델 만들기 : 입력층 - 은닉층1(256) - 은닉층2(128) - 은닉층3(64)- 출력층(10)

model = keras.models.Sequential([
      keras.layers.Flatten(input_shape = (28, 28)), #일열로 만들기
      #은닉층
      keras.layers.Dense(3000, activation='relu'), 
      keras.layers.Dense(3000, activation='relu'),
      keras.layers.Dense(3000, activation='relu'),
      keras.layers.Dense(10, activation='softmax'), # activation -> 확률을 출력
])


model.compile(optimizer= 'adam', loss = 'sparse_categorical_crossentropy', metrics=['accuracy'])


model.summary()
```

```python
#model 학습 시키기
model.fit(x_train, y_train,  epochs= 10)
```

```python
model.evaluate(x_test, y_test)
```
